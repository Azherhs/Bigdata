```python
%%file talkativecharacters.py

from mrjob.job import MRJob
from mrjob.step import MRStep
import os

class MRTalkativeCharacters(MRJob):

    def configure_args(self):
        super(MRTalkativeCharacters, self).configure_args()

    def mapper_init(self):
        # Extract episode number from file name
        try:
            self.episode = os.path.basename(os.path.splitext(os.environ['map_input_file'])[0])
        except KeyError:
            self.episode = 'Unknown'  # Fallback if episode is not set

    def mapper(self, _, line):
        fields = line.split('"')
        if len(fields) >= 5:
            character = fields[3]
            yield (self.episode, character), 1

    def combiner(self, key, values):
        yield key, sum(values)

    def reducer(self, key, values):
        yield None, (sum(values), key)  # Emit with a dummy key for sorting in the next step

    def final_reducer(self, _, values):
        sorted_values = sorted(values, reverse=True)  # Sort by line count in descending order
        for value in sorted_values:
            yield (value[1], value[0])  # Emit (Episode, Character), Count

    def steps(self):
        return [
            MRStep(mapper_init=self.mapper_init,
                   mapper=self.mapper,
                   combiner=self.combiner,
                   reducer=self.reducer),
            MRStep(reducer=self.final_reducer)
        ]

if __name__ == '__main__':
    MRTalkativeCharacters.run()
```

```python
%%file talkativecharacters2.py
from mrjob.job import MRJob
from mrjob.step import MRStep

class MRTalkativeCharacters(MRJob):
    
    def mapper(self, _, line):
        # Split the CSV formatted line
        fields = line.split('"')
        if len(fields) >= 5:
            character = fields[3]
            yield character, 1

    def combiner(self, character, counts):
        yield character, sum(counts)

    def reducer(self, character, counts):
        yield None, (sum(counts), character)

    # This secondary sort reducer will sort characters by the number of lines
    def reducer_sort(self, _, character_counts):
        for count, character in sorted(character_counts, reverse=True):
            yield character, count

    def steps(self):
        return [
            MRStep(mapper=self.mapper,
                   combiner=self.combiner,
                   reducer=self.reducer),
            MRStep(reducer=self.reducer_sort)
        ]

if __name__ == '__main__':
    MRTalkativeCharacters.run()
```

```python
%%file longestphrase.py

from mrjob.job import MRJob
from mrjob.step import MRStep
import re

class MRLongestPhrasePerCharacter(MRJob):

    def mapper(self, _, line):
        # Attempt to extract character and phrase using regex
        match = re.match(r'"[^"]+"\s+"([^"]+)"\s+"([^"]+)"', line)
        if match:
            character = match.group(1).strip()
            phrase = match.group(2).strip()
            if phrase:
                yield character, phrase

    def reducer(self, character, phrases):
        # Find the longest phrase for each character
        longest_phrase = max(phrases, key=len)
        yield character, longest_phrase

    def steps(self):
        return [MRStep(mapper=self.mapper, reducer=self.reducer)]

if __name__ == '__main__':
    MRLongestPhrasePerCharacter.run()
```
